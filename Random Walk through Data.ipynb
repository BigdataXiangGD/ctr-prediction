```python
train_df['hour_int'] = train_df.apply(lambda row: (row['hour'] - train_df['hour'].iloc[0]) / np.timedelta64(1, 'h'), axis=1)
```

```python
train_mod = train_df.drop('hour', axis=1)
```

```python
train_mod.info()
```

```python
features = ['banner_pos', 'site_category', 'device_id', 'device_type', 'device_conn_type', 'hour_int']
target = 'click'
```

```python
X = train_mod[features]
y = train_mod[target]
```
